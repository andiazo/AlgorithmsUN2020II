# 2. Apply the Classic and Po Shen Loh s methods for solving the following quadratic equations
*  x^2 = 0
*  (x-1)^2 = 0
*  x^2  1 =0
*  x^2+ 1 =0

Mtodos: Clsico y Po Shen Loh

```
# Classic Method
def classic(a, b, c):
  """ Mtodo clasico: Devuelve dos raices reales de una ecuacin cuadrtica de la forma ax^2 + bx + c """
  disc = b ** 2 - 4 * a * c
  if disc < 0:
    print('ERROR metodo clasico: La ecuacin no tiene soluciones reales')
    return 'NaN', 'NaN'
  elif disc == 0:
    x1 = -b / (2 * a)
    x2 = x1
    return x1, x2
  else:
    x1 = (-b - (disc ** 0.5)) / (2 * a)
    x2 = (-b + (disc ** 0.5)) / (2 * a)
  return x1, x2

# Po Shen Loh Method
def poShenLoh(a, b, c):
  """ Po-Shen Loh: Devuelve las raices reales de una ecuacin cuadrtica de la forma x^2 + bx + c """
  if a != 1:
    print("ERROR metodo Po Shen Loh: Este mtodo solo funciona cuando a = 1")
    return 'NaN', 'NaN'
  else:
    t = (-1*b) / 2
    if (((t**2) - c)<0):
      print('ERROR metodo Po Shen Loh: La ecuacin no tiene soluciones reales')
      return 'NaN','NaN'
    u = math.sqrt((t**2) - c)
    x1 = t - u
    x2 = t + u
  return x1, x2
```

```
# x^2 = 0
classol = classic(1,0,0)
posol = poShenLoh(1,0,0)
print('Las soluciones para x^2 = 0 son:')
print('Clasico: {} | Po Shen Loh: {} \n'.format(classol, posol))

# (x-1)^2 = (x-1)(x-1) = x^2 - 2x + 1 = 0
classol = classic(1,-2,1)
posol = poShenLoh(1,-2,1)
print('Las soluciones para x^2 - 2x + 1 = 0 son:')
print('Clasico: {} | Po Shen Loh: {} \n'.format(classol, posol))

# x^2  1 = 0
classol = classic(1,0,-1)
posol = poShenLoh(1,0,-1)
print('Las soluciones para x^2 - 1 = 0 son:')
print('Clasico: {} | Po Shen Loh: {}\n'.format(classol, posol))

# x^2 + 1 = 0
classol = classic(1,0,1)
posol = poShenLoh(1,0,1)
print('\nLas soluciones para x^2 + 1 = 0 son:')
print('Clasico: {} | Po Shen Loh: {}'.format(classol, posol))
```

## 3.1. z^2

```
def f(z):
    return z ** 2
 
def df(z):
    return 2 * z 
 
def newton_set1(z_c_r=0, z_c_i=0, zh = 5, zw=5, width=500, height=500, zoom=1,  niter=256, multiplicidad=1):
    """ Fractals using newton-raphson """
 
    # Pixels array
    pixels = np.arange(width*height*3, dtype=np.uint32).reshape(height, width, 3)
 
    h = 1e-7 # step size for numerical derivative
    eps = 1e-3 # max error allowed
 
    # Bounding roots
    r1 = 0
 
    # Color multiplication factor
    # NOTE: Increasing this darkens the image and lightens it otherwise
    multcol = 10
 
    for y in range(height):
            zy =   z_c_i+(((y-(height/2))/height)*zh*zoom)
 
            for x in range(width):
                zx = z_c_r+ (((x-(width/2))/width)*zw*zoom)
 
                # Mapping real -> complex plane
                z = complex(zx, zy)
                
                count = 0
                
                for i in range(niter):
                    # complex numerical derivative
                    dz = df(z)
                    if dz == 0:
                        break
 
                    count += 1
                    if count > 255:
                        break
                   
                    znext = z - multiplicidad*f(z) / dz # Newton iteration
 
                    if abs(znext - z) < eps: # stop when close enough to any root
                        break
                    
                    z = znext
 
                # Pixels colored using the roots
                if abs(z-r1)<eps:
                    # color red
                    pixels[height-y-1,x] = (255 - count*multcol, 0, 0)
                #elif abs(z-r2)<=eps:
                    # color green
                   # pixels[height-y-1,x] = (0, 255 - count*multcol, 0)
                else:   
                   pixels[height-y-1,x] = (0, 0, 0)
               
    return pixels
 
def display1(z_c_r=0, z_c_i=0, zh = 1, zw=1, width=500, height=500, zoom=1,  niter=256, multiplicidad=1, title="Newton Fractal f(x)"):
    """ Display a newton-ra uyphson fractal """
    print("Z region from (",z_c_r+ (((0-(width/2))/width)*zw*zoom),",",z_c_i+(((0-(height/2))/height)*zh*zoom) ,") to (",z_c_r+ (((width-(width/2))/width)*zw*zoom),",",z_c_i+(((height-(height/2))/height)*zh*zoom),")")
    pimg = newton_set1(z_c_r=z_c_r, z_c_i=z_c_i, zh = zh, zw=zw, width=width, height=width, zoom=zoom,  niter=niter, multiplicidad=multiplicidad)
    plt.title(title)
    plt.xlabel("real"+"("+ str(z_c_r+ (((0-(width/2))/width)*zw*zoom))+","+str(z_c_r+ (((width-(width/2))/width)*zw*zoom))+")")
    plt.ylabel("imaginary"+"("+str(z_c_i+(((0-(height/2))/height)*zh*zoom))+","+str(z_c_i+(((height-(height/2))/height)*zh*zoom)) +")")
    plt.imshow(pimg)
    plt.show()
 
titulo ="Newton Fractal f(z) = z^2"    
display1(zoom=1, multiplicidad=1, title=titulo)
display1(zoom=1/100, multiplicidad=1, title=titulo)
```
