# Plot of the recursive Fibonacci time [0,50] and the iterative Fibonacci time [0,500] including the line y=n to compare in a Colab notebook and upload the notebook in BubbleSortDistribution,ipynb in the group’s github

```
# import libraries
import matplotlib.pyplot as plt
import itertools as it
import numpy as np
import pandas as pd
import time
```

```
def fibR(n): # Recursive Fibonacci
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibR(n-1)+fibR(n-2)

def fib(n): # Iterative Fibonacci
    if n==0 :
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a+b
    return b
```

```
time_list = []
n_list = []
n=50
tstart = time.clock()
for i in range(n):
  ti = time.clock()
  fibR(i)
  tf = time.clock()
  time_list.append(tf-ti)
  n_list.append(i)
tend = time.clock()

x = np.array([i for i in range(n)])
y = x
plt.plot(x,y, label="y=n")
plt.plot(n_list,time_list, label="time complexity")
title = "Recursive Fibonacci [0,50] time complexity"
plt.title(title)
plt.xlabel("n")
plt.ylabel("seconds")
plt.legend()
plt.show

print("Run for:",(tend-tstart)/60,"minutes")
```

```
x = np.array([i for i in range(n)])
y = x*10
plt.plot(x,y, label="y=10*n")
plt.plot(n_list,time_list, label="time complexity")
title = "Recursive Fibonacci [0,50] time complexity"
plt.title(title)
plt.xlabel("n")
plt.ylabel("seconds")
plt.legend()
plt.show
```

```
time_list1 = []
n_list1 = []
n=500
for i in range(n):
  ti = time.clock()
  fib(i)
  tf = time.clock()
  time_list1.append(tf-ti)
  n_list1.append(i)

x = np.array([i for i in range(n)])
y = x
plt.plot(x,y, label="n")
plt.plot(n_list1,time_list1, label="time complexity")
title = "Iterative Fibonacci time complexity"
plt.title(title)
plt.xlabel("n")
plt.ylabel("seconds")
plt.legend()
plt.show
```
